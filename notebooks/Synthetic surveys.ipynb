```python
from eql_source_layouts import ground_survey, airborne_survey
import matplotlib.pyplot as plt
```

```python
region = (-50e3, 50e3, -50e3, 50e3)
```

```python
airborne = airborne_survey(region)
ground = ground_survey(region)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
tmp = ax.scatter(
    airborne.easting, airborne.northing, c=airborne.elevation, s=10
)
plt.colorbar(tmp, ax=ax)
ax.set_aspect("equal")
plt.show()

print("Number of points: {}".format(len(airborne)))
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
tmp = ax.scatter(
    ground.easting, ground.northing, c=ground.elevation, s=10
)
plt.colorbar(tmp, ax=ax)
ax.set_aspect("equal")
plt.show()

print("Number of points: {}".format(len(ground)))
```
